```python
df = pd.read_csv('/kaggle/input/music-education-dataset/music_education_dataset_new.csv')
```

```python
df_sorted = df.sort_values(by = 'Class_Level', ascending = False)
```

```python
df_sorted.head(10)
```

```python
# Check if there are any missing values
df.isnull().sum()
```

```python
# Replace the Class Level strings with integers
df['Class_Level'] = df['Class_Level'].replace({'Beginner' : 1, 'Intermediate' : 2, 'Advanced' : 3})
df
```

```python
# What's the average stress level based on class level?
df.groupby('Class_Level')['Stress_Level'].mean()
```

```python
# Count of each gender
df['Gender'].value_counts()
```

```python
# What is the average focus time for each gender?
df.groupby('Gender')['Focus_Time (min)'].mean()
```

```python
# What are the 10 youngest ages?
df['Age'].nsmallest(10)
```

```python
# What are the 10 oldest ages?
df['Age'].nlargest(10)
```

```python
# What is the average engagement score for age 13 or older and under age 13?
over_twelve = df['Age'] >= 13
df.groupby(over_twelve)['Engagement_Score'].mean().sort_values()
```

```python
# What is the average heart rate by instrument
df.groupby('Instrument')['Heart_Rate (BPM)'].mean().sort_values()
```

```python
# Lowest behavior scores
df['Behavior_Score'].nsmallest().sort_values()
```

```python
# Highest behavior Scores
df['Behavior_Score'].nlargest().sort_values()
```

```python
# Show Behavior score and gender columns, sort values in behavior column
df[(df['Behavior_Score'] >= 3)].sort_values(['Behavior_Score', 'Gender'], ascending = [True, False])[['Gender', 'Behavior_Score']]
```
